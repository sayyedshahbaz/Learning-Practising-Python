```python
import datetime
```

```python
today = datetime.datetime.today()
print(today)
```

```python
bday = datetime.datetime(1995, 10, 31, 19, 30, 15, 10)
print(bday)
```

```python
days_since_birth = (today - bday).days
print(days_since_birth)
```

```python
secs_since_birth = (today - bday).total_seconds()
print(secs_since_birth)
```

```python
#Add/subtract defined days from today
tdelta = datetime.timedelta(days=10)
print(today + tdelta)
```

```python
print(today.month)
print(today.day)
print(today.weekday())
print(today.year)
#Monday = 0 and Sunday = 6
```

```python
print(datetime.time(7, 2, 5, 15))
# datetime.date(Y, M, D)
# datetime.time(h, m, s, ms)
# datetime.datetime(Y, M, D, h, m, s, ms)
```

```python
#Add/Subtract defined hours from now
hour_delta = datetime.timedelta(hours=3)
print(datetime.datetime.now() + hour_delta)
```

```python
import pytz
```

```python
for tz in pytz.all_timezones:
    print(tz)
```

```python
datetime_today = datetime.datetime.now(tz=pytz.UTC)
print(datetime_today)
datetime_gmt = datetime_today.astimezone(pytz.timezone('GMT'))
print(datetime_gmt)
```

```python
#make datetime look nice
#strftime f = formatting
date_pretty = datetime_gmt.strftime('%B %d, %Y')
print(date_pretty)
```

```python
#nice to datetime object
#strptime p = parsing
date_ugly = datetime.datetime.strptime(date_pretty, '%B %d, %Y')
print(date_ugly)
```
